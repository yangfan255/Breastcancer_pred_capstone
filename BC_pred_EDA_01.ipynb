```python
corr = df.corr()
plt.figure(figsize = (10, 8))
sns.heatmap(corr)
plt.show()
```

```python
corr['diagnosis'].sort_values(ascending = False).to_frame('corr_value').reset_index()
```

```python
def find_high_correlated_features(frame):
    new_corr = frame.corr()
    new_corr.loc[:,:] = np.tril(new_corr, k=-1) # below main lower triangle of an array
    new_corr = new_corr.stack()
    print(new_corr[(new_corr > 0.4) | (new_corr < -0.4)])
    
find_high_correlated_features(df)
```
